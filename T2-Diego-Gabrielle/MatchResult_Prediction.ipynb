# Treinamento utilizando Arvore de Decisão

O primeiro algoritmo utilizado para realizar o treinamento foi o de Árvore de Decisão devido a sua facilidade de entendimento e, também, por ser conhecido como um dos mais eficientes quando se há classes e atributos bem definidos para realizar o treinamento. 

Basicamente o método funciona da seguinte maneira: Cria-se um nó raiz com a probabilidade de cada classe. Após isto é feita um busca dos nós a serem divididos, ou seja, cuja distribuição das probabilidades não classifique a amostra totalmente. Então é realizada a divisão destes nós disponíveis para divisão, utilizando um atributo que melhor classifica os dados (excluindo todos os atributos que inda não foram utilizados no caminho que começa deste o nó raiz até o nó a ser dividido). E, por fim, cria-se um novo nó com o atributo escolhido.

```python
from sklearn import tree
```

```python
classificador_tree = tree.DecisionTreeClassifier()
```

```python
classificador_tree = classificador_tree.fit(data_train,resultado_target)
```

```python
result_pred = classificador_tree.predict(data_train)
```

```python
result_pred
```

```python
resultado_target
```

# Avaliação da Arvore de Decisão

```python
from sklearn import metrics
metrics.accuracy_score(resultado_target,result_pred)
```

```python
metrics.precision_score(resultado_target,result_pred,average='macro')
```

# Validação Cruzada da Árvore de Decisão

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(classificador_tree,data_train,resultado_target,cv=None,verbose=3,scoring='accuracy')
```

```python
np.mean(scores)
```

# Treinamento utilizando Random Forest

Para termos uma "segunda opinião", optamos por realizar o treinamento utilizando o algoritmo de Random Forest.
Optamos por este, pois utiliza vários subconjuntos de árvores dedecisão para fazer a classificação.

Funcionamento: Para classificar um novo objeto com base em atributos, cada árvore dá uma classificação e dizemos a árvore “vote” para essa classe. A floresta escolhe a classificação que tenha a maioria dos votos (mais de todas as árvores da floresta).

```python
from sklearn.ensemble import RandomForestClassifier
classificador_random = RandomForestClassifier(max_depth=6,random_state=0,min_samples_split=3)
classificador_random.fit(data_train,resultado_target)
#resultado_target
```

```python
data_pred = classificador_random.predict(data_train)
data_pred
```

# Avaliação do Random Forest

```python
metrics.accuracy_score(resultado_target,data_pred)
```

# Validação Cruzada do Random Forest

```python
scores = cross_val_score(classificador_random,data_train,resultado_target,cv=None,verbose=10,scoring='accuracy')
np.mean(scores)
```

Cria-se uma nova voluna para comparar o resultado predito com a coluna RESULTADO

```python
info_2016_pred = info_2016
info_2016_pred["RESULTADO_PREDITO"] = data_pred

Resultado_pred = {0: "Vitória_Mandante", 1: "Empate", 2: "Vitória_Visitante"}

coluna_resultado_pred = info_2016_pred["RESULTADO_PREDITO"].apply(Resultado_pred.get)

info_2016_pred["RESULTADO_PREDITO"] = coluna_resultado_pred
info_2016_pred["RESULTADO"] = coluna_resultado.apply(Resultado_pred.get)
info_2016_pred

info_2016_pred.to_csv('Avaliacao_RandomForesto.csv')
info_2016_pred
```
